```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
not_optimized_times = [641, 657, 641, 625, 657, 640, 641, 719, 719, 687, 750] 
optimized_times = [688, 609, 578, 516, 578, 594, 515, 547, 515, 516, 500] 

sum_not_optimized_times = 0
sum_optimized_times = 0

optimized_entries= 0
not_optimized_entries = 0

average_optimized = 0
average_not_optimized = 0

for value in not_optimized_times:
    sum_not_optimized_times += value
    not_optimized_entries += 1

for value in optimized_times:
    sum_optimized_times += value
    optimized_entries += 1

average_not_optimized = sum_not_optimized_times / not_optimized_entries
average_optimized = sum_optimized_times / optimized_entries
print(f'Durchschnittszeit nicht optimiert: {average_not_optimized}')
print(f'Durchschnittszeit optimiert: {average_optimized}')
```

```python
all_data = [not_optimized_times, optimized_times]

fig = plt.figure(figsize =(10, 7))
 
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(all_data, patch_artist=True)
ax.set_xticklabels(['Vor der Optimierung', 'Nach der Optimierung'])
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.set_ylabel("Zeit in Millisekunden")
 
plt.savefig('../visualisations/IOOptimzation.png', bbox_inches='tight')
```
